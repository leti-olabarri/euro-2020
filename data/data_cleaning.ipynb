```python
def str_to_int(percentage):
    percentage = int(percentage.replace("%", ""))
    return percentage

columns_to_change = ["possession_home", "possession_away", "duels_won_home", "duels_won_away"]

for i in columns_to_change:
    df[i] = df[i].apply(str_to_int)
```
