```python
def str_to_int(percentage):
    percentage = int(percentage.replace("%", ""))
    return percentage

columns_to_change = ["possession_home", "possession_away", "duels_won_home", "duels_won_away"]

for i in columns_to_change:
    df[i] = df[i].apply(str_to_int)
```

De momento, me voy a quedar con esta información. Voy a quitar las tres últimas columnas (`events_list`, `lineup_home` y `lineup_away`), porque de momento no voy a usar esos datos. Si los uso, será de otra manera. Voy a añadir una columna vacía, `venue_id`, por si me da tiempo a hacer geoqueries con los estadios en los que se jugó cada partido.

```python
df = df.drop(columns=["events_list", "lineup_home", "lineup_away"])
```

```python
len(df.index)
```

```python
venues = []
for i in range(0,51):
    venues.append(None)
len(venues)
```

```python
df["venue_id"] = venues
```

```python
df.head()
```

```python
df["team_name_home"] = df["team_name_home"].apply(lambda x: x.strip())
df["team_name_away"] = df["team_name_away"].apply(lambda x: x.strip())
```

```python
df["stage"] = df["stage"].apply(lambda x: x.strip())
```

```python
df["team_name_home"][0]
```

```python
df["match_id"] = df.index
df["match_id"] = df["match_id"].apply(lambda x: x+1)
df.head()
```

```python
user = os.getenv("POSTGRES_USER")
password = os.getenv("POSTGRES_PASSWORD")
```

```python
postgres_url = f"postgresql+psycopg2://{user}:{password}@127.0.0.1:5432/euro2020"
```

```python
engine = create_engine(postgres_url)
euro2020 = engine.connect()
euro2020
```

```python
df.to_sql("match", euro2020, index=False, if_exists="replace")
```
